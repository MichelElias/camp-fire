# Overview


## Summary

The Camp Fire was perhaps the most destructive recorded wildfire in California's history and was one of the world's most destructive natural disasters in 2018.  All totaled, it burned over 153,000 acres and destroyed over 13,500 structures [Camp Fire](https://en.wikipedia.org/wiki/Camp_Fire_(2018)) and [Cal Fire Incident](http://cdfdata.fire.ca.gov/incidents/incidents_details_newsreleases?incident_id=2277).  The fire began on November 8, 2018 and was not 100% contained until November 25, 2018.  During that time, the event was a social media and news topic of substantial interest, and it was widely covered on both media platforms.

The work in this notebook covers techniques for processing text data related to that event.  It does so by employing three tool sets to analyze tweet and news information specifically related to the Camp Fire with the goal of using these tools to categorize individual elements into groups that can be used in further analysis.  The inputs to the steps described here are tweets and news stories and the outputs are labels for each tweet and story placing that item in a group with other similar items.


## Process

The goal of this analysis is to label an individual tweet or news story as belonging to a category of similar items.  Therefore the steps described here are done to produce a DataFrame containing columns of labels alongside rows of tweets or news articles.  In addition, a summary DataFrame aggregates (by count) this output at 12-hour time windows, 07:00 and 19:00 local time each day, during the course of the fire.  These times correspond with the release of Cal Fire's [incident reports](http://cdfdata.fire.ca.gov/incidents/incidents_details_newsreleases?incident_id=2277).

The steps followed to achieve this objective were:

1. Read in Twitter and News data
2. Conduct a Latent Dirichlet Allocation (LDA) analysis on each tweet and story
3. Conduct a Singular Value Decomposition (SVD) and K-Means clustering analysis on each tweet and story
4. Conduct a Doc2Vec and K-Means clustering analysis on each tweet and story
5. Append new columns representing the labels resulting from steps 2, 3 and 4 to DataFrames
6. Count the frequency of these labels occurring in each 12-hour window
7. Save the data to a Structured Query Language (SQL) database and to Comma Separated Value (CSV) files


## Data

The text data used in this analysis were collected from two Application Programming Interfaces (API).  It should be noted that in both cases, the data were pulled using queries designed to return results related to the Camp Fire.  Search criteria include keywords and specific data ranges corresponding in November 2018.  This approach has the advantage of narrowing the dataset to information that is likely to be related to the event but makes the text categorization task more challenging since the documents are examination likely to be similar.

* Twitter API: Twitter's [Premium Search API](https://developer.twitter.com/en/docs/tweets/search/api-reference/premium-search) was used to pull tweet data with the following parameters:
    * keywords: campfire,
    * Language: English
    * Profile Region: California
    * Excluding retweets
    * Start time: November 08, 2018 02:00 UTC 
    * End time: November 26, 2018 02:00 UTC

* News API: The [News API](https://newsapi.org/) was used to pull news stories with the following parameters:
    * keywords: wildfire, camp fire, percent contained, evacuate, flames, paradise, cal fire,
    * Start Date: 2018-11-08
    * End Date: 2018-11-30
    
Due to data limitations, we have opted to use the entire dataset to train the categorization models.  This should be considered when using the results for further analysis because it may introduce some error due to variance.  The problem arises from information availability since the categorization of a tweet or news story at the beginning of the fire may use information from items only available at the end of the event.  Therefore, the resulting categorization may not have been possible at the beginning of the incident and should be cautiously used in predicted modeling.  Extending this approach to additional similar events will help build a dataset that could be used for predictive modeling and possibly in the a real-time tool that can be employed as part of an emergency response.

## Detailed steps

The following provide more detail into the steps followed in this notebook:

1. Read in data
    * Read in tweet data 
        * Use clean_text column in which words are lower case and punctuation and URLs have been removed
    * Read in news data
        * Scrub the text field by setting to lower case and removing punctuation and URLs
2. Examine tweets
    * Conduct the LDA analysis
        * Vectorize the clean_text field using a Count Vectorizer and remove stop words
        * Build the LDA model specifying the number of topics (10)
        * Examine the word rankings in each topic
        * Output a scores DataFrame showing the scoring of each tweet for each topic
        * Select the maximum score and assign that topic to the tweet
        * Append the scored labels to the original tweets data frame
        * Create a DataFrame of topic counts by 12-hour window over the fire's existence
        * Plot the count of each tweet topic group over the course of the fire
    * Conduct the SVDC analysis
        * Vectorize the clean_text field using a Term Frequency - Inverse Document Frequency (tfidf) Vectorizer and remove stop words
        * Build the SVD decomposition specifying the number of dimensions (100)
        * Examine the word rankings in each dimension
        * Output a scores DataFrame showing the mapping of each tweet for each dimension
        * Use K-Means specifying the number of clusters (10) to group each tweet along SVD dimensions
        * Append the assigned cluster labels to the original tweets data frame
        * Create a DataFrame of topic counts by 12-hour window over the fire's existence        
        * Plot the count of each tweet cluster over the course of the fire
    * Conduct the d2vc analysis
        * Prepare sentences by creating tagged documents for each tweet in which stop words have been removed
        * Build the doc2vec model specifying the resulting vector dimensions (100)
        * Generate vector representations for each tweet
        * Use K-Means clustering specifying the number of groups (10) to group the vectors together by cosine similarity distance
        * Append the assigned cluster labels to the original tweets data frame
        * Create a DataFrame of topic counts by 12-hour window over the fire's existence        
        * Plot the count of each tweet cluster over the course of the fire
3. Examine news     
    * Conduct the LDA analysis
        * Follow the LDA steps outlined for tweets using the clean_content field
    * Conduct the SVDC analysis
        * Follow the SVDC steps outlined for tweets using the clean_content field
    * Conduct the d2vc analysis
        * Follow the d2vc steps outlined for tweets using the clean_content field
4. Output the results
    * Create a single DataFrame of topic and cluster counts by 12-hour window over the fire's existence
    * Write the updated tweet and news DataFrames and the count DataFrame to SQL tables
    * Save these DataFrames as CSV files

## Establish classes

```python
class DocumentTagging(object):
    def __init__(self, doc_series):
        self.doc_series = doc_series
    def __iter__(self):
        for i, doc in enumerate(self.doc_series):
            yield TaggedDocument(words = doc.split(), tags = [i])
            
```

## Establish functions

```python
# Create a function to scrub a text string
def scrub_text(in_text):
    # Remove urls, numbers and punctuation
    word_list = re.sub(r'(http)\S*|[^a-zA-Z 0-9\n]','',in_text.lower()).split()
    return ' '.join(word_list)


# Return a DataFrame of the count vectorized words
def clean_text_vectorize(text_ser = None, vec_model = 'count', ngram_range = (1,1), use_stop_words = True):
    # Check the data type of ser
    if not isinstance(text_ser, pd.Series):
        raise Exception('Input text_ser must be a pandas Series')

    # Identify the stop words
    if use_stop_words:
        stop_words = list(ENGLISH_STOP_WORDS)
        stop_words.extend(['http','amp', 'campfire'])
    else:
        stop_words = None

    # set up a vectorizer, cv for count, tfidf for tfidf
    if vec_model == 'count':
        cv = CountVectorizer(min_df = 2, 
                             ngram_range = ngram_range, 
                             stop_words = stop_words)
    elif vec_model == 'tfidf':
        cv = TfidfVectorizer(min_df = 2, 
                         ngram_range = ngram_range, 
                         stop_words = stop_words)
    else:
        raise Exception('Unrecognized value for vec_model')
        
    # return the sparse word vector
    vec_sparse = cv.fit_transform(text_ser)

    # convert to a data frame and return
    return pd.DataFrame(vec_sparse.toarray(), 
                        columns = cv.get_feature_names())


# Put words and columns of lda output into a DataFrame
def data_frame_components(components = None, columns = None):
    df_components = []
    for i in range(len(components)):
        # For this component, get the word scores sorted by score and return a list of strings
        word_score = [str(round(sc, 1)) + ": " + wd for sc, wd in \
                      sorted(zip(components[i], columns), reverse = True)]
        # Append this to a list
        df_components.append(word_score)
        
    # Convert to a DataFrame and transpose    
    return pd.DataFrame(df_components, index = add_component_names(len(components))).T


# Create a scoring matrix with component scores and individual
def create_score_frame(fit_model = None, append_labels = True, 
                       label = 'c', vec_df = None):
    
    # Create a scores matrix by transforming the
    score_array = fit_model.transform(vec_df)
    
    # Convert the array to a DataFrame
    scores = pd.DataFrame(data = score_array,
                         columns = add_component_names(score_array.shape[1], label))
    
    if append_labels:
        # Add columns for the sum, max and label
        scores['sum'] = scores.iloc[:,0:score_array.shape[1]].sum(axis=1)
        scores['max'] = scores.iloc[:,0:score_array.shape[1]].max(axis=1)
        scores['label'] = scores.iloc[:,0:score_array.shape[1]].idxmax(axis=1)
    
    return scores


# Add labels to the original dataframe
def add_observation_labels(df = None, col_name = 'label_x', labels = None):
    # Create a new column with labels
    df[col_name] = labels


#Add component names
def add_component_names(n_components = None, label = 'c'):
    # Create a list of component names
    return [f'{label}_{i}' for i in range(n_components)]


# Build a lda model
def build_decom_model(vec_df = None, decom_model = 'lda', 
                      n_components = 10):
    # Check the data type of ser
    if not isinstance(vec_df, pd.DataFrame):
        raise Exception('Input vec_df must be a pandas DataFrame')

    # Instantiate and fit a lda model
    if decom_model == 'lda':
        lda = LatentDirichletAllocation(n_components = n_components)
    elif decom_model == 'svd':
        lda = TruncatedSVD(n_components = n_components)
    else:
        raise Exception('Unrecognized value for vec_model')        
        
    lda.fit_transform(vec_df)
    
    return lda


# Draw a plot of the columns of a dataframe
def df_line_plot(df = None, title = 'title', x_label = 'x_label', 
                 y_label = 'y_label', figsize = (8,8)):
    
    # Establish figure size
    fig, ax = plt.subplots(figsize = figsize)
    
    # plot the lines
    labels = []
    for col in df.columns:
        ax = sns.lineplot(x=df.index, y = df[col], ax = ax)
#         ax = sns.barplot(x=df.index, y = df[col], ax = ax)
        labels.append(col)
        
    # Set the title, x_label and y_label
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    
    # put the legend
    if len(labels) < 30:
        ax.legend(labels, loc = 1)
     
    # Set the x axis labels
    ax.set_xticklabels(df.index, rotation = 90)

    
# Plot the explained variance from the component analysis
def exp_var_plot(decom_obj = None, title = 'title', x_label = 'x_label', 
                 y_label = 'y_label', figsize = (8,8)):
    
    # Establish figure size
    fig, ax = plt.subplots(figsize = figsize)
    
    labels = []
    # Calculate the x range
    x = np.array(range(len(decom_obj.explained_variance_ratio_)))
    labels.append('Explained variance')
    
    # Plot the explained variance for each component
    ax = sns.barplot(x = x, y = decom_obj.explained_variance_ratio_, ax = ax)
    labels.append('Cumulative variance')
    
    # Plot the cumulative for each component
    ax = sns.lineplot(x = x, y = np.cumsum(decom_obj.explained_variance_ratio_), ax = ax) 
    
    # Set the title, x_label and y_label
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_xticks(np.arange(0, len(x), 5))
    ax.set_xticklabels(np.arange(0, len(x), 5))
    ax.set_ylabel(y_label)
    
    # put the legend
    ax.legend(labels, loc = 2)
    

# Build a kmeans clustering model and return the fitted model object
def build_kmeans_model(data_df = None, n_clusters = 3):
    # Check the data type of DataFrame
    if not isinstance(data_df, pd.DataFrame):
        raise Exception('Input data_df must be a pandas DataFrame')    
    
    # Instantiate and fit a kmeans models
    km = KMeans(n_clusters = n_clusters, random_state = 42)
    km.fit(data_df)
    
    # return the fitted model
    return km


def build_kmeanscluster_model(data_vectors = None, n_clusters = 3):    
    kc = KMeansClusterer(num_means = n_clusters, 
                         distance = nltk.cluster.util.cosine_distance, 
                         repeats = 3)
    assigned_clusters = kc.cluster(data_vectors, assign_clusters=True)
    return assigned_clusters


# Add text to the cluster labels of a fitted model
def cluster_labels(fit_model = None, label = 'D2'):
    # return label names
    return [f'{label}_{i}' for i in fit_model.labels_]

# Add text to the cluster labels of a fitted model
def cluster_labels_kmc(assigned_clusters = None, label = 'D2'):
    # return label names
    return [f'{label}_{i}' for i in assigned_clusters]
 
    
# Write the posts data to a SQL table
def write_to_database(df, engine = None, table = None, index = False, if_exists = 'append'):
    # write posts to the posts table
    if engine == None:
        return
    else:
        df.to_sql(table, con=engine, index=index, if_exists=if_exists)

        
# Read in data from SQL database
def read_from_database(SQL, engine = None):
    # write posts to the posts table
    if engine == None:
        return
    return pd.read_sql(SQL, con = engine)


# Build a doc2vec model
def build_d2v_model(tagged_docs = None, vector_size = 50, 
                    min_count= 2, alpha = 0.025, epochs = 20,
                    model_file = None):
    # Build the doc2vec model
    model = Doc2Vec(vector_size = vector_size, min_count=min_count, alpha=alpha, epochs=epochs)
    # Build a vocabulary
    model.build_vocab(tagged_docs)
    # Train the model
    model.train(tagged_docs,total_examples=model.corpus_count, epochs=model.epochs)
    # Save the model
    if model_file != None:
        model.save(model_file)
    
    return model


# Generate doc2vec scores from a fitted model
def generate_d2v_scores(d2v_fit_model = None, tag_docs = None):
    # output a scores matrix
    vectors = [d2v_fit_model.infer_vector(doc.words) for doc in tag_docs]
    col_names = [f'a_{i}' for i in range(d2v_fit_model.vector_size)]

    return pd.DataFrame(data = vectors, columns = col_names)


# Generate doc2vec scores from a fitted model
def generate_d2v_vectors(d2v_fit_model = None, tag_docs = None):
    # output a scores matrix
    vectors = [d2v_fit_model.infer_vector(doc.words) for doc in tag_docs]

    return vectors


# Pull key words from text
def key_words(in_text):
    stop_words = list(ENGLISH_STOP_WORDS)
    stop_words.extend(['http','amp', 'campfire'])
    word_list = [word for word in in_text.split() if word not in stop_words]
    return ' '.join(word_list)

```
